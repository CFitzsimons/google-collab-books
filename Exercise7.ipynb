```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/suzannelittle/ca682i/master/data/sampler/DataSampler.csv')
```

```python
df.head()
```

```python
df['A'] = pd.to_datetime(df['A'])
```

```python
# df['B'].min()
df.dtypes
```

```python
f,ax = plt.subplots()
ax1 = df.plot.line(x='A', y='B', color="skyblue", ax=ax)
ax1.legend(loc='center', bbox_to_anchor=(0.42, 1.05), frameon=False)
ax1.set_yticks([0, 0.25, 0.5, 0.75, 1])
ax1.set_ylim(0, 1)
ax1.set_ylabel("B (random float)")
ax1.set_xlabel("Timestamp")

ax1.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d %H:%M:%S'))

ax1.grid(visible=True, axis='both', which='both', linestyle='-')

# f.autofmt_xdate()

ax2 = ax1.twinx()
ax2.set_yticks([0, 25, 50, 75, 100])
ax2.set_ylim(0, 100)
ax2.set_ylabel("C (random integeger between 2 & 88)")

final = df.plot.line(x='A', y='C', ax=ax2, color="red")
final.legend(loc='center', bbox_to_anchor=(0.58, 1.05), frameon=False)
```
